```python
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
sns.set_theme(style="whitegrid")

data_path = "/Users/khali/Desktop/Coding Workspace/Flight Delay Prediction/data/raw/Airline_Delay_Cause.csv"
df=pd.read_csv(data_path)

print("Shape: " , df.shape , "Rows and Columns")
display(df.head())

df["delay_rate"] = df["arr_del15"] / df["arr_flights"]

df["date"] = pd.to_datetime(df[["year" , "month"]].assign(day=1)) #so this gets converted to like then date time thing


plt.figure(figsize=(10,5))
sns.boxplot(data=df, x = "carrier" , y = "delay_rate")
plt.title("Delay Rate Distribution by Carrier")
plt.ylabel("Delay Rate")
plt.xlabel("Carrier")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
print("\nMissing Values: \n" , df.isnull().sum()) #no missing values

processed_Path = "/Users/khali/Desktop/Coding Workspace/Flight Delay Prediction/data/processed/Airline_Delay_Cause_Clean.csv"
df.to_csv(processed_Path , index = False)
print(f"\nCleaned data, along with Delay_Rate and Date is saved to: {processed_Path}")
```
